# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# parmetros
mu = 15       # mdia
sigma = 0.1   # desvio padro
x = 15.2      # valor observado

# calcular o z-score
z = (x - mu) / sigma

# calcular a probabilidade de X > 15.2
prob = 1 - stats.norm.cdf(z)

print(f"A probabilidade de um pacote pesar mais que 15,2 kg  {prob:.4f}")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# parmetros
mu = 15       # mdia
sigma = 0.1   # desvio padro

# calcular o percentil de 95%
percentil_95 = stats.norm.ppf(0.95, loc=mu, scale=sigma)

print(f"O percentil de 95%  {percentil_95:.4f} kg")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui


# distribuio normal padro (mdia=0, desvio=1)
prob_normal = stats.norm.cdf(Y)

# distribuio t-student com 5 graus de liberdade
prob_t = stats.t.cdf(Y, df=5)

# Exemplo: exibir os 5 primeiros valores de cada
print("Primeiros valores - Normal padro:")
print(prob_normal[:5])

print("\nPrimeiros valores - t-Student (df=5):")
print(prob_t[:5])
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# vetor de valores
Y = np.linspace(-6, 6, 200)

# calcular probabilidades acumuladas
prob_normal = stats.norm.cdf(Y)
prob_t = stats.t.cdf(Y, df=5)

# plot
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_normal, label='Normal Padro', color='blue')
plt.plot(Y, prob_t, label='t-Student (df=5)', color='orange', linestyle='--')
plt.title('Funes de Distribuio Acumulada (CDF)')
plt.xlabel('Valor')
plt.ylabel('Probabilidade acumulada')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y = np.linspace(0, 1, 20)

# calcular os percentis (quantis)
percentis_normal = stats.norm.ppf(Y)
percentis_t = stats.t.ppf(Y, df=5)

# Exibir os primeiros valores
print("Percentis - Normal Padro:")
print(percentis_normal[:5])

print("\nPercentis - t-Student (df=5):")
print(percentis_t[:5])
```

6. Faa um grfico com os dados do item anterior.

```python
# vetor de probabilidades
P = np.linspace(0, 1, 20)

# calcular os percentis (inversa da CDF)
percentis_normal = stats.norm.ppf(P)
percentis_t = stats.t.ppf(P, df=5)

# plotando
plt.figure(figsize=(10, 6))
plt.plot(P, percentis_normal, label='Normal Padro', color='blue')
plt.plot(P, percentis_t, label='t-Student (df=5)', color='orange', linestyle='--')
plt.title('Comparao de Percentis: Normal vs t-Student (df=5)')
plt.xlabel('Probabilidade acumulada (p)')
plt.ylabel('Valor correspondente (percentil)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
