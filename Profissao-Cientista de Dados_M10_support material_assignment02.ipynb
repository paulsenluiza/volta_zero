1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Carregar os dados
df = pd.read_csv("previsao_de_renda.csv")

# Criar boxplots para visualizar a distribuição da renda com relação às variáveis categóricas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['posse_de_veiculo'], y=df['renda'])
plt.title("Renda x Posse de Veículo")

plt.subplot(1, 2, 2)
sns.boxplot(x=df['posse_de_imovel'], y=df['renda'])
plt.title("Renda x Posse de Imóvel")

plt.tight_layout()
plt.show()

# Teste t para posse de veículo
renda_veiculo_sim = df[df['posse_de_veiculo'] == True]['renda']
renda_veiculo_nao = df[df['posse_de_veiculo'] == False]['renda']
t_stat_veiculo, p_valor_veiculo = ttest_ind(renda_veiculo_sim, renda_veiculo_nao, equal_var=False)

# Teste t para posse de imóvel
renda_imovel_sim = df[df['posse_de_imovel'] == True]['renda']
renda_imovel_nao = df[df['posse_de_imovel'] == False]['renda']
t_stat_imovel, p_valor_imovel = ttest_ind(renda_imovel_sim, renda_imovel_nao, equal_var=False)

# Exibir os resultados
print(f"Teste t para posse de veículo: p-valor = {p_valor_veiculo:.10f}")
print(f"Teste t para posse de imóvel: p-valor = {p_valor_imovel:.10f}")

# Interpretação:
if p_valor_veiculo < 0.05:
    print("Posse de veículo discrimina a renda de forma estatisticamente significante.")
else:
    print("Posse de veículo NÃO discrimina a renda de forma estatisticamente significante.")

if p_valor_imovel < 0.05:
    print("Posse de imóvel discrimina a renda de forma estatisticamente significante.")
else:
    print("Posse de imóvel NÃO discrimina a renda de forma estatisticamente significante.")
````
